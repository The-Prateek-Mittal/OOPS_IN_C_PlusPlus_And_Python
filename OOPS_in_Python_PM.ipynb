```python
# The _init_ method 

class computer : 
    def __init__(self,cpu,ram) : # kind of constructor 
        self.cpu = cpu 
        self.ram = ram 
        
        
    def config(self):
        print("CPU :", self.cpu ," Ram : ", self.ram)
    
com1 = computer('15',16) # decides the size of class 
com2 = computer('Ryzen3',8)
com1.config()
com2.config()

# self is used to differentiate between two or more objects of the same class. 
```

```python
# Comparing objects of same class 

class computer : 
    def __init__(self,cpu,ram) : # kind of constructor 
        self.cpu = cpu 
        self.ram = ram 
        
        
    def config(self):
        print("CPU :", self.cpu ," Ram : ", self.ram)
    
    def compare(self, other):
        if self.ram == other.ram :
            return True 
        else :
            return False
    
com1 = computer('15',16) # decides the size of class 
com2 = computer('Ryzen3',8)
if com1.compare(com2) :
    print("They are same.")
else :
    print("They are different.")
```

```python
# They are two types of variable : instance variable and static (class) variable.

class car : 
    wheels = 4  # Static Variable 
    def __init__(self) : 
        self.mil = 10
        self.com = "BMW"  # com and mil are instance variables
        
c1 = car()
c2 = car()

c1.mil = 8 

print(c1.com,c1.mil,c1.wheels)
print(c2.com, c2.mil,c2.wheels)
car.wheels = 5 # Updating the static variable.
print(c1.com,c1.mil,c1.wheels)
print(c2.com, c2.mil,c2.wheels)
```

```python
# Types of Methods : Instance, Class, Static 

class student :
    school = "Telusko"
    def __init__(self,m1,m2,m3):
        self.m1 = m1
        self.m2  = m2
        self.m3 = m3
    
    def avg(self):  # Instance method - Two types - Accessors(getters) and Mutators (setters)
        return (self.m1 + self.m2 + self.m3)/3
    
    @classmethod # makes the class method (decorator)
    def info(cls):  # Class method 
        return cls.school
    
    @staticmethod #(decorator)
    def info2(): # No self, no cls as it is a static method.
        print("This is a student class")
        
s1 = student(45,46,47)
s2 = student(89,32,12)
print(s1.avg())
print(student.info())
student.info2()
```

```python
# Constructor in inheritence and method resolution order.

class A : 
    def __init__(self):
        print("in A init")
        
    def feature(self):
        print("Feature 1 working")
        
    def feature2(self):
        print("Feature 2 working")

class B(A) : 
    def __init__(self):
        super().__init__() # Calling both the constructors. 
        print("in B init")
    
    def feature3(self):
        print("Feature 3 working")
        
    def feature4(self):
        print("Feature 4 working")
        
a1 = B()



```

```python
# MRO 
class A : 
    def __init__(self):
        print("in A init")
        
    def feature(self):
        print("Feature 1 working")
        
    def feature2(self):
        print("Feature 2 working")

class B() : 
    def __init__(self):
        super().__init__() # Calling both the constructors. 
        print("in B init")

    def feature3(self):
        print("Feature 3 working")
        
    def feature4(self):
        print("Feature 4 working")

class C(A,B):
    def __init__(self):
        super().__init__() # Calling both the constructors. 
        print("in C init")

    def feature5(self):
        print("Feature 5 working")
        

a1 = C() # Why it is calling A and C not B and C or A,B and C ?
# It will prefer left one. (This is for methods also)
    
    
```

```python
# Polymorphism 

# Duck Typing in Python 

# Duck Test : if features of a bird is matching with the duck so it is a duck 
class Pycharm :
    def execute(self):
        print("Compiling")
        print("Runnning")
        
class myeditor :
    def execute(self):
        print("Spell Check")
        print("Convention Check")
        
        
class laptop :
    def code(self, ide) :
        ide.execute()

ide = Pycharm()
lap1 = laptop()
lap1.code(ide)
ide1 = myeditor()
lap1.code(ide1)
```

```python
# Operator Overloading 

class Student :
    def __init__(self,m1, m2):
        self.m1  = m1
        self.m2 = m2
    
    def __add__(self, other):
        m1 = self.m1 + other.m1
        m2 = self.m2 + other.m2
        m3 = Student(m1,m2)
        return m3 
    

s1 = Student(58,69)
s2 = Student(60, 65)
s3 = s1 + s2

print(s3.m1)

```

```python
# Method Overloading : Creating two or more methods in the same class but with same name and different number of arguments. 

class Student :
    def __init__(self,m1, m2):
        self.m1  = m1
        self.m2 = m2
        
    def sum(self, a=None, b=None,c=None ) :
        if a!= None and b!= None and c != None :
            s = a+b+c 
            return s
        elif a!= None and b != None :
            s = a+b 
            return s 
        else :
            return a 

s1 = Student(58,69)
print(s1.sum(2,3))
```
